Problem 1: Student Records Challenge

Given a nested list of student records, write code to:

students = [
    ["Alice", "Physics", 85],
    ["Bob", "Chemistry", 78],
    ["Cara", "Maths", 92]
]

a. Print just the names and scores of all students.
b. Add a new student record ["Dave", "Biology", 88] at the end.
c. Update Bob's score to 82.

```python
students = [
    ["Alice", "Physics", 85],
    ["Bob", "Chemistry", 78],
    ["Cara", "Maths", 92]
]

# a. Print just the names and scores of all students
for student in students:
    print(student[0], student[2])

# b. Add a new student record ["Dave", "Biology", 88] at the end
students.append(["Dave", "Biology", 88])
print(students)

# c. Update Bob's score to 82
for student in students:
    if student[0] == "Bob":
        student[2] = 82
print(students)
```

Problem 2: Modify a Sensor Readings List

Modify a Sensor Readings List
You have a list of readings from a temperature sensor every hour:

readings = [23, 25, 24, 22, 26, 27, 28, 29]

a. The sensor misreported the third and fourth hours: replace readings at indices 2 and 3 with [24, 23].
b. Append a final reading of 30.
c. Use a list method to find how many times the value 24 appears.

```python
readings = [23, 25, 24, 22, 26, 27, 28, 29]

# a. Replace readings at indices 2 and 3 with [24, 23]
readings[2:4] = [24, 23]
print(readings)

# b. Append a final reading of 30
readings.append(30)
print(readings)

# c. Find how many times the value 24 appears
print(readings.count(24))
```

Problem 3: String Slicing & Indexing - Device ID Parser

String Slicing & Indexing - Device ID Parser
Given a device ID string:

device_id = "SN-A1B2C3D4E5"

a. Extract the serial prefix ("SN-") and suffix ("A1B2C3D4E5") using slicing.
b. Retrieve just the last three characters ("4E5") using negative indexing.
c. Reverse the entire string using slicing with a negative step

```python
device_id = "SN-A1B2C3D4E5"

# a. Extract the serial prefix ("SN-") and suffix ("A1B2C3D4E5") using slicing
prefix = device_id[:3]
suffix = device_id[3:]
print("prefix: ", prefix, ", suffix: ", suffix)

# b. Retrieve just the last three characters ("4E5") using negative indexing
last_three = device_id[-3:]
print(last_three)

# c. Reverse the entire string using slicing with a negative step
reversed_id = device_id[::-1]
print(reversed_id)
```

Problem 5: Combined Loop + List Operation Challenge

Combined Loop + List Operation Challenge
Given a list of mixed data types representing readings and status flags:

data = [100, "OK", 105, "FAIL", 102, "OK", 99, "OK"]

a. Loop through the list and print only the numeric readings.
b. Use list slicing to create two new lists: one for all even indices, one at odd indices.
c. Count how many "OK" flags appear.

```python
data = [100, "OK", 105, "FAIL", 102, "OK", 99, "OK"]

# a. Loop through the list and print only the numeric readings
for item in data:
    if type(item) == int:
        print(item)

# b. Use list slicing to create two new lists: one for all even indices, one at odd indices
even_indices = data[::2]
odd_indices = data[1::2]
print(even_indices, odd_indices)

# c. Count how many "OK" flags appear
ok_count = data.count("OK")
print(ok_count)
```
