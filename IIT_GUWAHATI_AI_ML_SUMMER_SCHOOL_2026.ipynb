**Problem 2:**

Simple Calculator (with Safety Check) Problem: Create a calculator that performs addition, subtraction, multiplication, or division.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operation (+, -, *, /): ")

if operator == '+':
    print("Result:", num1 + num2)
elif operator == '-':
    print("Result:", num1 - num2)
elif operator == '*':
    print("Result:", num1 * num2)
elif operator == '/':
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Division by zero is not allowed")
else:
    print("Invalid operator")
```

**Problem 6:**

Sum of First N Natural Numbers Problem: Calculate the sum of the first n natural numbers

```python
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total += i
print("Sum of first", n, "natural numbers is:", total)
```

**Problem 7:**

Multiplication Table Problem: Print the multiplication table of a given number.

```python
number = int(input("Enter a number: "))
for i in range(1, 11):
    print(number, "x", i, "=", number * i)
```

**Problem 8:**

Count Positive, Negative, and Zero Problem: Take 5 numbers from the user and count positives, negatives, and zeros.

```python
positive = negative = zero = 0
for i in range(5):
    num = int(input("Enter number: "))
    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1
    else:
        zero += 1
print("Positive:", positive)
print("Negative:", negative)
print("Zero:", zero)
```
